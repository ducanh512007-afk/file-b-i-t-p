# 1
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để
tính tống các số nguyên từ 1 đến n. (n <= 1000) 


```python
n = int(input())
x = 0 # kho lưu trữ
for i in range (1, n + 1): # for: duyệt qua tất cả các phần tử trong 1 range
    x += i # Cộng dồn
print(f'Tong tu 1 đến {n} là {x}')
```

```python
n = int(input())
x = 0
i = 1
while i <= n: # while: lặp lệnh đến khi điều kiện sai
    x += i
    i += 1
print(f'Tổng từ 1 đến {n} là {x}')
```

# 8
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n = int(input())
i = 0

def số_đảo(n):
    a = str(n)
    a = str(n)[::-1]
    return int(a)
while True:
    n = n + số_đảo(n)
    i += 1
    if số_đảo(n) == n: 
        print(f'Với {i} bước, có số đối xứng {n}')
        break
  
```

# 10
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất


đề cần: 
- lập dãy collatz
- in x với độ dài tương ứng
- in  x sao cho nó bé nhất với L lớn nhất
tư duy: 
- lập def collatx cho dễ
  - dãy collatz: list giá trị cho đến khi n = 1. 
  -> tạo 1 list tí nữa tìm max L  
  điều kiện: giá trị cuối =1 -> dùng while n khác 1
- in ra độ dài max với i bé nhất: 
  - tìm trong khoảng -> dùng range
  - lắp biến L = chiều dài list
  - so   

```python
n = int(input())
b = 0 # Max L
c = 0 # số i nhỏ nhất có max L

def a(n):  # Lập hàm Dãy Collatz
    x = []
    x.append(n)
    while n != 1: 
        if n % 2 == 0: 
            n = n/2
        else: 
            n = (3 * n) + 1
        x.append(n)
    return x 
    
 

# Dùng vòng for để so sánh liên tục các giá trị L xem giá trị nào lớn nhất. 
for i in range(1, n + 1):
  
    L = len(a(i)) # Độ dài dãy L
    
    if L > b: # để L > b thì lệnh while chạm đúng giá trị i đầu tiên có L max. 
        b = L 
        c = i 
        # Logic: Quét hết range, lấy giá trị số i và max L
print(f'Số {c} có dãy collatz dài nhất với L = {b}')
```

Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220 
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false" 


```python
a = int(input())
b = int(input())
c = 0
d = 0
for i in range(1, a): 
    if a % i == 0:
        c += i
for k in range(1, b): 
    if b % k == 0: 
        d += k
if c == b and d == a:
    print('True')
else:
    print('False')


```
