# 22
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 


```python
n = int(input())
s = str(n)
a = 0
b = 0
for i in s[::1]: 
    if int(i) % 2 == 0:
        a += 1
    else: 
        b += 1
print(f'có {a} số chẵn, {b} số lẻ')
```

```python
n = int(input())
b = 0
c = 0
while n > 0:
    a = n % 10
    if a % 2 == 0:
        b += 1
    else:
        c += 1
    n //= 10
print(b,c)
        
```

# 23

Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 

```python
n = int(input())
s = 0
k = 0
while s + (k + 1) < n: 
    k += 1
    s += k
print(f' Tong la {s}')
```

# 24: 
Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 


```python
A = int(input())
k = 1 
s = 1
while s + (1/k) <= A:
    k += 1
    s += 1/k
print(k+1)
```
