Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.


```python
from math import*
a = float(input())
b = float(input())
c = float(input())
if a + b < c and a + c < b and b + c < a: print('Khong la ba canh cua mot tam giac')
else:
    if a==b==c: print('đều')
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2: print('vuong')
    elif a==b or a==c or b==c: print('can')
    else: print('thuong')
    p = (a + b + c)/2
    s = sqrt(p * (p-a) * (p-b) * (p-c))
    y = a + b + c
print(f'{s:.2f}, {y}')

```
