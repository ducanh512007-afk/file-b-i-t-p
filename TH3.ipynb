# W3A1
Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.



```python
n = int(input("Nhập số nguyên n: "))
if 100 <= n <= 999: 
    x = n % 10
    y = (n//10) % 10
    z = n // 100
a = x * 100 + y * 10 + z
print(a)
```

# w3A3
Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2
hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0


```python
n = int(input())
if n & (n - 1) == 0: 
    print('True')
else: 
    print('False')
```

```python
m = float(input())
n = float(input())
if m/n > 0:
    print(-(-m//n))
if m/n < 0: 
    print(-(-m//n))
```
