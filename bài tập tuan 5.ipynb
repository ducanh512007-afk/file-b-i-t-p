int:

```python
a, b = map(int, input().split())
def Max_Of_Two( a, b):
    if a > b: 
        return a 
    else:
        return b
print(Max_Of_Two( a, b))
```

Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]: 

```python
a, b = map(int, input().split())
def swap(a, b):
    a, b = b, a
    return a, b
print(tuple(swap(a, b)))
```

VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

```python
a = int(input())
def so_hoan_hao(a):
    b = 0
    for i in range(1, a):
        if a % i == 0:
            b += i
        if b == a:
            return'So hoan hao'
            break
    else:
        return'Ko phai so hoan hao'
print(so_hoan_hao(a))
                
                
            
        
```

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1

```python
a = list(map(int, input().split()))
k = int(input())
def quet_so(a, k):
    for i in a:
        if i == k:
            return i
            break 
    else:
        return -1

        

print(quet_so(a, k))
```

# W5A6
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 

```python
n = int(input())
def giai_thua(n):
    for i in range(n-1, 0, -1):
        n *= i
    return n
print(giai_thua(n))
```

# W5A8
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.

```python
x, y = map(int,input().split())
def hamming(x, y):
    a = 0

    n  =x^y
    while n > 0:
        if n % 2 == 1:
            a += 1
            n //= 2
        else:
            n //= 2
    return a
print(hamming(x, y))

        


    
```

# W5A9
Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
n = int(input())
def tong_chu_so(n):
    b = 0
    while n > 0:
        a = n % 10
        b += a
        n //= 10
    return b
print(tong_chu_so(n))

```

# W5A10
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài.

```python
a, b = input().split()
def dang_cau(a, b):
    for i in range(0, len(a)):
        for k in range(0, len(a)):
            if (a[i] == a[k] and b[i] != b[k]) or (a[i] != a[k] and b[i] == b[k]):
                return'false'
                break
    else: 
        return'true'
print(dang_cau(a, b))

    
            

```

# W5A11
Viết hàm is_product_divisible_by_10(a, b) trả về True nếu tích a*b chia hết cho 10, ngược lại False

```python
a, b = map(int, input().split())
def is_product_divisible_by_10(a, b):
    n = a*b
    if n % 10 == 0:
        return'True'
    else:
        return'False'
print(is_product_divisible_by_10(a, b))
```

# W5A15
 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python
a, b ,c =map(int, input().split())
def is_right_triangle(a, b, c):
    if a + b > c and a + c > b and b + c > a:
        if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
            return'True'
        else:
            return'False'
    else:
        return'False'
print(is_right_triangle(a, b, c))
    
```
